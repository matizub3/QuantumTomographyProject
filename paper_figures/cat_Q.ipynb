```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import jax
import jax.numpy as jnp
```

```python
#load ../baselines/loss_evolution/cgan_losses.dat
with open("../baselines/loss_evolution/cgan_losses.dat", "rb") as f:
    cgan_losses = pickle.load(f)

#load ../baselines/loss_evolution/imle_losses.dat
with open("../baselines/loss_evolution/imle_losses.dat", "rb") as f:
    imle_losses = pickle.load(f)

#load ../loss_evolution_results/CNF_cat_FFJORD_SINE_Adaptive_L1_uniform_cosine_decay.dat
with open("../loss_evolution_results/Q/CNF_cat_FFJORD_SINE_Adaptive_L1_uniform_cosine_decay.dat", "rb") as f:
    CNF_losses = pickle.load(f)
```

```python
epochs = np.arange(0, 500, 10)

plt.plot(epochs, CNF_losses["L1_uniform"][::10], label = "CNF L1", color = "tab:blue", linestyle = "-")
plt.plot(epochs, CNF_losses["KL_uniform"][::10], label = "CNF KL", color = "tab:orange", linestyle = "-")
plt.plot(epochs, CNF_losses["KL_rev_uniform"][::10], label = "CNF KL rev", color = "tab:green", linestyle = "-")

epochs = np.arange(0, 501, 10)
plt.plot(epochs, imle_losses["L1"][::10], label = "IMLE L1", color = "tab:blue", linestyle = "--")
plt.plot(epochs, imle_losses["KL"][::10], label = "IMLE KL", color = "tab:orange", linestyle = "--")
plt.plot(epochs, imle_losses["KL_rev"][::10], label = "IMLE KL rev", color = "tab:green", linestyle = "--")

epochs = np.arange(0, 500, 10)
plt.plot(epochs, cgan_losses["L1"][::10], label = "CGAN L1", color = "tab:blue", linestyle = ":")
plt.plot(epochs, cgan_losses["KL"][::10], label = "CGAN KL", color = "tab:orange", linestyle = ":")
plt.plot(epochs, cgan_losses["KL_rev"][::10], label = "CGAN KL rev", color = "tab:green", linestyle = ":")

plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
```
